```python
import numpy as np
import pandas as pd
import torch
torch.set_default_tensor_type(torch.FloatTensor)
```

```python
# 线性回归
x = np.random.normal(0,5,[1000,3])
bete = np.matrix([10,20,30])
y = x.dot(beta.T) + 3 + 0.1 * np.random.normal(0,1,[1000,1])
```

```python
# 转换为tensor, 并且将类型转为float，torch中将float64类型认为是double
x_data = torch.from_numpy(x)
y_data = torch.from_numpy(y)
x_data = x_data.float()
y_data = y_data.float()
```

```python
class LinearModel(torch.nn.Module):
    def __init__(self):
        super(LinearModel,self).__init__()
        self.linear = torch.nn.Linear(3,1)
    def forward(self,x):
        y_pred = self.linear(x)
        return y_pred
```

```python
model = LinearModel()
criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(),lr = 0.00001)
for epoch in range(1000):
    y_pred = model(x_data)
    loss = criterion(y_pred,y_data)
    print("------------------")
    print(epoch,loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print(model.linear.weight)
print(model.linear.bias)
```
